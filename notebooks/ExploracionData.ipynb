## Ingenieria de Caracteristicas

Se explorara la opcion de crear nuevas variables que capturen mejor la variable target compra de seguro al siguiente mes.

Posibles nuevas variables:


- Ratio de utilizaciÃ³n de la tarjeta (VAR015 ): VAR04 / VAR02.
- Promedio de pago mensual (VAR016): VAR09 / 6.
- Ãndice de morosidad (VAR017): VAR05 + VAR06.
- Estabilidad financiera (VAR018): VAR07 / VAR09.


```python
df_Procesada["VAR015"] = np.where(df_Procesada["VAR02"] != 0, df_Procesada["VAR04"] / df_Procesada["VAR02"], 0)
df_Procesada["VAR016"] = df_Procesada["VAR09"] / 6
df_Procesada["VAR017"] = df_Procesada["VAR05"] + df_Procesada["VAR06"]
df_Procesada["VAR018"] = np.where(df_Procesada["VAR09"] != 0, df_Procesada["VAR07"] / df_Procesada["VAR09"], 0)
```
